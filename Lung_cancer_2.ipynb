```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'C:\Users\USER\Desktop\Python class\Dataset\Apps and Scripts Technology (DS Dataset)\Lung Cancer\survey lung cancer.csv')
```

```python
data = data.drop_duplicates()
data.info()
```

```python
from sklearn import preprocessing
label = preprocessing.LabelEncoder()
data['GENDER'] = label.fit_transform(data['GENDER'])
data['LUNG_CANCER'] = label.fit_transform(data['LUNG_CANCER'])
data.corr()
```

```python
plt.figure(figsize = (12,8))
sns.heatmap(data.corr(), annot = True, cmap = 'Blues')
```

```python
data['LUNG_CANCER'].value_counts()
```

```python
X = data.iloc[:,0:-1]
y = data.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 10)
```

```python
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
logr = LogisticRegression()
```

```python
logr.fit(X_train, y_train)
```

```python
y_pred = logr.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(confusion_matrix(y_pred,y_test))
```

```python
print(classification_report(y_pred,y_test))
```

```python
print(round(accuracy_score(y_pred,y_test),2))
```

```python
from imblearn.over_sampling import ADASYN
```

```python
adasyn = ADASYN(sampling_strategy='auto',random_state = 2,n_neighbors=7)
X_resampled,y_resampled = adasyn.fit_resample(X,y)
```

```python
y_resampled.value_counts()
```

```python
X_train_resampled, X_test_resampled, y_train_resampled, y_test_resampled =train_test_split(X_resampled,y_resampled,test_size= 0.2, random_state = 5 )
```

```python
print(X_train_resampled.shape,X_test_resampled.shape,y_train_resampled.shape,y_test_resampled.shape)
```

```python
logr.fit(X_train_resampled,y_train_resampled)
```

```python
y_pred_resampled = logr.predict(X_test_resampled)
```

```python
print(y_pred_resampled)
```

```python
Confused_matrix = confusion_matrix(y_pred_resampled,y_test_resampled)
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(Confused_matrix, annot = True, cmap = 'Greens')
```

```python
print(classification_report(y_pred_resampled,y_test_resampled))
```

```python
print(round(accuracy_score(y_pred_resampled,y_test_resampled),2))
```

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train_resampled,y_train_resampled)
```

```python
y_pred_decision_tree = decision_tree.predict(X_test_resampled)
```

```python
print(classification_report(y_pred_decision_tree,y_test_resampled))
```

```python
Confused_decision_matrix = confusion_matrix(y_pred_decision_tree,y_test_resampled)
```

```python
print(Confused_decision_matrix)
```

```python
plt.figure(figsize= (12,8))
sns.heatmap(Confused_decision_matrix, annot = True, cmap = 'Blues')
```
